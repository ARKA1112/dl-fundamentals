```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt


relu = nn.ReLU()


inputs = torch.arange(-3, 4)
outputs = relu(inputs)

fig = plt.figure(figsize=(5.5, 3))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.431))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.plot(inputs.numpy(), outputs.numpy())

inputs = torch.arange(-3, 4, 0.001)
outputs = torch.where(inputs >= 0., 1., 0.)
plt.plot(inputs.numpy(), outputs.numpy())


plt.savefig('relu-derivative.pdf')
plt.show()
```

```python
gelu = nn.GELU()


inputs = torch.arange(-4, 4, 0.1).float()
outputs = gelu(inputs)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.506))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.plot(inputs.numpy(), outputs.numpy())

plt.savefig('gelu.pdf')
plt.show()
```

```python
from torch.autograd import grad

gelu = nn.GELU()


inputs = torch.arange(-4, 4, 0.1).float()
outputs = gelu(inputs)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.506))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(inputs.numpy(), outputs.numpy())

inputs = torch.arange(-4, 4, 0.1, requires_grad=True).float()
outputs = gelu(inputs)
grad(outputs, inputs)
plt.plot(inputs.numpy(), outputs.numpy())


plt.savefig('gelu-deriv.pdf')
plt.show()
```

```python
from torch.autograd import grad

inputs = torch.arange(-4, 4, 0.1, requires_grad=True).float()
outputs = gelu(inputs)
grad(outputs, inputs)
```

```python
sigmoid = nn.Sigmoid()


inputs = torch.arange(-7, 7, 0.1).float()
outputs = sigmoid(inputs)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.506))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.plot(inputs.numpy(), outputs.numpy())

plt.savefig('sigmoid.pdf')
plt.show()
```

```python
sigmoid = nn.Sigmoid()


inputs = torch.arange(-7, 7, 0.1).float()
outputs = sigmoid(inputs)* (1-sigmoid(inputs))

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.504))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.ylim([0, 1])

plt.plot(inputs.numpy(), outputs.numpy(), color='C1')

plt.savefig('sigmoid-deriv.pdf')
plt.show()
```

```python


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('axes', 0.506))
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

sigmoid = nn.GELU()
inputs = torch.arange(-7, 7, 0.1).float()
outputs = sigmoid(inputs)
plt.plot(inputs.numpy(), outputs.numpy(), label='GELU (nn.GELU)')

sigmoid = nn.SiLU()
inputs = torch.arange(-7, 7, 0.1).float()
outputs = sigmoid(inputs)
plt.plot(inputs.numpy(), outputs.numpy(), label='Swish (nn.SiLU)')

sigmoid = nn.Mish()
inputs = torch.arange(-7, 7, 0.1).float()
outputs = sigmoid(inputs)
plt.plot(inputs.numpy(), outputs.numpy(), label='Mish (nn.Mish)')

plt.legend()
plt.savefig('trends.pdf')
plt.show()
```
